## Описание данных

**Таблица `books`**

Содержит данные о книгах:

- `book_id` — идентификатор книги;
- `author_id` — идентификатор автора;
- `title` — название книги;
- `num_pages` — количество страниц;
- `publication_date` — дата публикации книги;
- `publisher_id` — идентификатор издателя.

**Таблица `authors`**

Содержит данные об авторах:

- `author_id` — идентификатор автора;
- `author` — имя автора.

**Таблица `publishers`**

Содержит данные об издательствах:

- `publisher_id` — идентификатор издательства;
- `publisher` — название издательства;

**Таблица `ratings`**

Содержит данные о пользовательских оценках книг:

- `rating_id` — идентификатор оценки;
- `book_id` — идентификатор книги;
- `username` — имя пользователя, оставившего оценку;
- `rating` — оценка книги.

**Таблица `reviews`**

Содержит данные о пользовательских обзорах на книги:

- `review_id` — идентификатор обзора;
- `book_id` — идентификатор книги;
- `username` — имя пользователя, написавшего обзор;
- `text` — текст обзора.

## Задачи

### Посчитайте, сколько книг вышло после 1 января 2000 года

```python
query = '''
SELECT COUNT(book_id) AS books_published
FROM books
WHERE publication_date > '2000-01-01'
'''
con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

После 1 января 2000 года вышло 819 книг, то есть за 20 лет - это, примерно, по 41 книге в год

### Для каждой книги посчитайте количество обзоров и среднюю оценку

```python
query = '''
SELECT 
--b.book_id AS book_id,
title,
COUNT(DISTINCT r.review_id) AS reviews_cnt,
AVG(ra.rating) AS avg_rating
FROM books as b
LEFT JOIN reviews AS r on b.book_id = r.book_id
LEFT JOIN ratings AS ra on b.book_id = ra.book_id
GROUP BY b.book_id
ORDER BY reviews_cnt, avg_rating DESC
'''
con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

Самые высокие оценки у книг, на которые нет обзоров, самые низкие у книг с наибольшим количеством обзоров. Возможно, это обусловлено тем, что недовольные читатели чаще хотят объяснить низкую оценку и дают отзыв, довольные просто ставят оценку. Сумерки лидируют по количеству обзоров, но имеют низкий рейтинг.

### Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры;

```python
query = '''
SELECT 
pb.publisher,
tab.published_books
FROM
(
SELECT
publisher_id,
COUNT(book_id) AS published_books
FROM books
WHERE book_id in
(SELECT
book_id
FROM books
WHERE num_pages > 50)
GROUP BY publisher_id
ORDER BY published_books DESC
LIMIT 1
) AS tab
LEFT JOIN publishers AS pb ON tab.publisher_id = pb.publisher_id
'''
con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

Издательство Penguin Books выпустило наибольшее число книг толще 50 страниц

### Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками;

```python
query = '''
SELECT
author,
ROUND(avg_rating) AS avg_rating
FROM
(
    SELECT
    b.author_id,
    AVG(tab.avg_rating) AS avg_rating
    FROM
(
        SELECT
        book_id,
        AVG(rating) AS avg_rating,
        COUNT(rating_id) AS rating_cnt
        FROM ratings
        GROUP BY book_id
        HAVING COUNT(rating_id) >= 50
        ORDER BY avg_rating DESC
) AS tab
    LEFT JOIN books AS b ON tab.book_id = b.book_id
    GROUP BY b.author_id
    ORDER BY avg_rating DESC
    LIMIT 1
) AS tab2
LEFT JOIN authors AS a ON tab2.author_id = a.author_id


'''
con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

Книги J.K. Rowling о Гарри Поттере имеют самую высокую среднюю оценку среди книг с 50 и более оценками. Это неудивительно: эти книги популярны по всему миру.

### Посчитайте среднее количество обзоров от пользователей, которые поставили больше 48 оценок.

```python
query = '''
SELECT
AVG(review_cnt) AS avg_review_cnt

FROM
(
    SELECT 
    username,
    COUNT(review_id) AS review_cnt
    FROM reviews
    WHERE username IN
(
        SELECT
        username
        FROM ratings
        GROUP BY username
        HAVING COUNT(rating_id) > 48
)
    GROUP BY username
    ) AS tab
   
'''
con=engine.connect()

pd.io.sql.read_sql(sql=text(query), con = con)
```

В среднем пользователи, которые поставили больше 48 оценок, оставляют 24 обзора.
